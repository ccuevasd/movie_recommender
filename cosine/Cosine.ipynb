```python
users = R.index
```

```python
R.index
```

```python
 #OPTION 2 KNN
# Fill in missing values
# Instantiate Imputer
imputer = KNNImputer(n_neighbors = 5)
```

```python
# Create Dataframe 
R_imputed = pd.DataFrame(imputer.fit_transform(R), columns= R.columns, index=users)
R_imputed
```

```python
# cosim_matrix
similarities = pd.DataFrame(cosine_similarity(R_imputed), columns=users, index=users)
similarities
```

```python
# Make a heatmapt out of the cosim_matrix
sns.heatmap(similarities)
```

```python
# Introduce New User
# Calculate the cosine similarity with all of them
Dummy = {'Toy Story (1995)': 5, 'Sabrina (1995)': 1, 'My Man Godfrey (1957)': 2} # onely rates two movies
Dummy = pd.DataFrame(Dummy, index=['Dummy'])
Dummy
```

```python
# Append Paul to the user-item matrix
R_new_user = R_imputed.append(Dummy)
R_new_user
```

```python
# Fill the NaNs for Paul
R_new_user_filled = R_new_user.fillna(2.5)
R_new_user_filled
```

```python
### Create a filter for the missing movies
# ~is basically saying "not": turns the boolean values around
movie_filter = ~R_new_user.isna().any().values
movie_filter
```

```python
# Create an updated user list
updated_users = R_new_user.index
updated_users
```

```python
R_new_user.transpose()[movie_filter].transpose()
```

```python
# calculate a similarity to the other users based on dummys rating
similarities_new_user = pd.DataFrame(cosine_similarity(R_new_user.transpose()[movie_filter].transpose()), \
                                    index=updated_users, columns=updated_users)
similarities_new_user
```

```python
# Predict ratings for Dummy
similarities_Dummy = similarities_new_user['Dummy'][~(similarities_new_user.index=='Dummy')]
similarities_Dummy
```

```python
# Calculate rating predictions
rating_predictions = pd.DataFrame(\
                        np.dot(similarities_Dummy, R_imputed) \
                        /similarities_Dummy.sum(), \
                        index=R.columns)
rating_predictions
```

```python
# Get recommendations
rating_predictions[~movie_filter].sort_values(by=0, ascending=False)
```
